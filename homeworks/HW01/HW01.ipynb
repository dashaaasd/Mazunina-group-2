Поэлементное сложение

```python
aa+ab
```

Поэлементное умножение

```python
aa*ab
```

Умножение на скаляр

```python
aa*2
```

Среднее значение

```python
np.mean(aa)
```

Сумма всех элементов

```python
np.sum(aa)
```

Минимальное, максимальное значение

```python
np.min(aa)
```

```python
np.max(aa)
```

```python
np.std(aa)
```
